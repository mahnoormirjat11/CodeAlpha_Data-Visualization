```python
# Load the uploaded Kaggle dataset for heart disease
import pandas as pd

# Load dataset
file_path = "Heart_Disease_ (1).csv"
df = pd.read_csv(file_path)

# Display first few rows to understand the structure
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a consistent visual style
sns.set(style="whitegrid")

# 1. Bar chart: Count of patients with and without heart disease
plt.figure(figsize=(7, 5))
sns.countplot(x='HeartDisease', data=df, palette='Set2')
plt.title("Heart Disease Count (0 = No, 1 = Yes)")
plt.xlabel("Heart Disease")
plt.ylabel("Number of People")
plt.tight_layout()
plt.show()

# 2. Boxplot: Cholesterol levels by heart disease status
plt.figure(figsize=(7, 5))
sns.boxplot(x='HeartDisease', y='Cholesterol', data=df, palette='coolwarm')
plt.title("Cholesterol Levels vs Heart Disease")
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("Cholesterol (mg/dL)")
plt.tight_layout()
plt.show()

# 3. Scatter Plot: Age vs Max Heart Rate by Heart Disease
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Age', y='MaxHeartRate', hue='HeartDisease', palette='coolwarm', s=100)
plt.title("Age vs Max Heart Rate by Heart Disease")
plt.xlabel("Age")
plt.ylabel("Max Heart Rate")
plt.legend(title='Heart Disease')
plt.tight_layout()
plt.show()

# 4. Correlation Heatmap (for numerical columns)
plt.figure(figsize=(10, 8))
corr_matrix = df.select_dtypes(include=['int64', 'float64']).corr()
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu', fmt=".2f", square=True)
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.show()
```
